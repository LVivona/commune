## What is c.Module

### Reusable

c.Module is a collection of functions that can be reusable for many applicaitons. This is where you would want to put functions in that you do not have to reimport, or would require you to find the code, etc. The purpose of this desing is to enable faster reusability without reimplementation. 

### Non-Oppinionated

The c.Module is intended to be Non-Opppinionated to prevent vendor locking. This is obviously not perfect, but it a goal we contineu to strive for. We do this by having the developer choose what functions they want to reuse and replace.
